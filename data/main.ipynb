```python
import pandas as pd
import matplotlib.pyplot as plt
import spacy
```

```python
data = pd.read_excel('processed_medication_data.xlsx')
route_counts = data['Route of administration'].value_counts()
data.head(100)
```

```python

total_rows = data.shape[0]


print("Total number of rows:", total_rows)
```

```python
unique_routes_count = data['Route of administration'].nunique()


print("Number of different routes of administration:", unique_routes_count)
```

```python

route_counts = data['Route of administration'].value_counts()

route_counts = route_counts.head(15)


plt.figure(figsize=(15, 7))
route_counts.plot(kind='bar')
plt.xlabel('Route of Administration')
plt.ylabel('Count')
plt.title('Medicinal Drugs by Route of Administration')
plt.xticks(rotation=90)


plt.tight_layout()
plt.show()
```

```python
#40 substâncias ativas mais comuns
substance_counts = data['Active substance\n'].value_counts()


top_40_substances = substance_counts.head(40)


plt.figure(figsize=(12, 6))  
ax = top_40_substances.plot(kind='bar')
ax.set_xlabel('Active substance\n')
ax.set_ylabel('Count')
ax.set_title('Top 40 Medicinal Drugs by Active substance')
ax.tick_params(axis='x', rotation=90)


plt.tight_layout()
plt.show()
```

## Gráficos da metadata recolhida

```python

holder_counts = data['Marketing authorisation holder'].value_counts().head(20)


plt.figure(figsize=(10, 10))
plt.pie(holder_counts, labels=holder_counts.index, autopct='%1.1f%%', startangle=140)
plt.axis('equal') 


plt.title("Top 25 Distribution Marketing Authorisation Holder")


plt.show()
```

```python
column_name = 'Genérico:'

# Get value counts for the specified column
value_counts = data[column_name].value_counts()

# Select the top N values (for example, top 10)
top_N = 10
top_values = value_counts.head(top_N)

# Create a bar chart
plt.figure(figsize=(12, 6))
top_values.plot(kind='bar')
column_name = 'Genérico'
plt.title(f'Values in Column {column_name}')
plt.xlabel(column_name)
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Specify the column name
column_name = 'Duração do Tratamento:'

# Get value counts for the specified column and select the top 10
top_10_values = data[column_name].value_counts().head(10)

# Create a bar chart for the top 10 values
plt.figure(figsize=(12, 6))
top_10_values.plot(kind='bar')
column_name = 'Duração do Tratamento'
plt.title(f'Values in Column {column_name}')
plt.xlabel(column_name)
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
