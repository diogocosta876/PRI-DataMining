```python
import pandas as pd
import matplotlib.pyplot as plt
import spacy
```

```python
data = pd.read_excel('processed_medication_data.xlsx')
route_counts = data['Route of administration'].value_counts()
data.head(100)
```

```python
# Count the total number of rows
total_rows = data.shape[0]

# Print the result
print("Total number of rows:", total_rows)
```

```python
unique_routes_count = data['Route of administration'].nunique()

# Display the count
print("Number of different routes of administration:", unique_routes_count)
```

```python
# Count the occurrences of each route of administration
route_counts = data['Route of administration'].value_counts()

route_counts = route_counts.head(15)

# Create a bar graph
plt.figure(figsize=(15, 7))
route_counts.plot(kind='bar')
plt.xlabel('Route of Administration')
plt.ylabel('Count')
plt.title('Medicinal Drugs by Route of Administration')
plt.xticks(rotation=90)

# Show the graph
plt.tight_layout()
plt.show()
```

```python
#40 substâncias ativas mais comuns

substance_counts = data['Active substance\n'].value_counts()

# Select the top 40 most common substances
top_40_substances = substance_counts.head(40)

# Create a bar graph for the top 40 substances
plt.figure(figsize=(12, 6))  # Adjust the figure size
ax = top_40_substances.plot(kind='bar')
ax.set_xlabel('Active substance\n')
ax.set_ylabel('Count')
ax.set_title('Top 40 Medicinal Drugs by Active substance')
ax.tick_params(axis='x', rotation=90)

# Show the graph
plt.tight_layout()
plt.show()
```

```python
# Function to calculate the average text length for a column
def average_text_length(column):
    return column.apply(lambda x: len(str(x)) if isinstance(x, str) else 0).mean()

# Calculate the average text length for each column that contains text data
avg_text_lengths = data.select_dtypes(include=['object']).apply(average_text_length)

# Print or save the average text lengths for each column
print(avg_text_lengths)
# avg_text_lengths.to_excel("average_text_lengths.xlsx", index=False)  # Uncomment to save to an Excel file
```

```python
# Load the spaCy NER model for English
nlp = spacy.load("en_core_web_sm")


# Function to apply NER to a text and return the entities
def apply_ner(text):
    doc = nlp(text)
    entities = [(ent.text, ent.label_) for ent in doc.ents]
    return entities

# Function to check if text is medium-sized
def is_medium_text(text, max_length=200):
    return len(str(text)) <= max_length

# Iterate through the entire DataFrame and apply NER to medium-sized text
for column in data.columns:
    if data[column].dtype == 'object':  # Check if the column contains text data
        data['Entities_' + column] = data[column].apply(lambda x: apply_ner(x) if is_medium_text(x) else [])

# Print or save the updated data with NER results
print(data)
# data.to_excel("data_with_ner.xlsx", index=False)  # Uncomment to save to an Excel file
```
